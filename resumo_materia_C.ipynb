## 1. Fundamentos e Tipos Básicos
Alguns dos tipos de dados mais comuns em C incluem:
- `int`: para números inteiros.
- `float`: para números de ponto flutuante (decimais).
- `double`: para números de ponto flutuante com maior precisão.
- `char`: para caracteres individuais ou strings.

No printf das variaveis, alguns pontos importantes a se lembrar:
- Para imprimir um `int`, usamos `%d`.
- Para imprimir um `float`, usamos `%f`.
- Para imprimir um `char`, usamos `%c`.
- Para imprimir uma `string`, usamos `%s`.
- Quando definimos uma `string`, devemos usar aspas duplas, e para um `char`, usamos aspas simples.


```c
#include <stdio.h>

int main(void) {
    //como fazer a definição das variaveis
    int idade = 20;
    float altura = 1.75;
    char inicial = 'A';
    char nome[] = "André"; // A capacidade de armazenamento da string vai ser definida o do conteudo dentro da string
    //neste caso, vai ser 5
    char nome[10] = "André"; // A capacidade aqui é definida pelo o numero dentro dos parenteses retos
    
    printf("Inicial: %c\n", inicial);
    printf("Nome: %s\n", nome);
    printf("Idade: %d anos\n", idade);
    printf("Altura: %.2f metros\n", altura);

    
    return 0;
}
```

##### Coisas a ter atenção na criação dos nomes para as variáveis:
- Os nomes devem começar com uma letra ou um sublinhado (`_`).
- Os nomes podem conter letras, dígitos e sublinhados, mas não podem conter espaços ou caracteres especiais.
- Os nomes não podem ser palavras reservadas da linguagem C (como `int`, `float`, `if`, `while`, etc.).
- Os nomes devem ser descritivos para facilitar a leitura do código, por exemplo, `contador` em vez de `c`.

##### Exemplos de variaveis válidas:
- `times10`
- `get_Next_Char`
- `_done`

##### Exemplos de variaveis inválidas:
- `10times` (não pode começar com um dígito)
- `get Next Char` (não pode conter espaços)
- `get-Next-Char` (não pode conter caracteres especiais como `-`)
- `int` (não pode ser uma palavra reservada)
- `máximo` (não pode conter caracteres acentuados)


## 2. Expressões e Operadores
O C suporta vários operadores aritméticos (`+`, `-`, `*`, `/`, `%`), relacionais (`>`, `<`, `==`, `!=`) e lógicos (`&&`, `||`, `!`). Cuidado com a divisão entre inteiros, pois o resultado será um número inteiro (ex: `3 / 4` dá `0`).

Operadores aritméticos:
- `+` : Adição. Exemplo: `3 + 4 = 7`.
- `-` : Subtração. Exemplo: `5 - 2 = 3`.
- `*` : Multiplicação. Exemplo: `2 * 3 = 6`.
- `/` : Divisão. Exemplo: `6 / 2 = 3`.
- `%` : Módulo (resto da divisão). Exemplo: `5 % 2 = 1`.

```c
#include <stdio.h>

int main(void) {
    int a = 10, b = 3;
    
    printf("Soma: %d\n", a + b);
    printf("Divisao inteira: %d\n", a / b);
    printf("Resto da divisao (modulo): %d\n", a % b);
    
    return 0;
}
```

## 3. Estruturas Condicionais
As estruturas condicionais permitem executar blocos de código diferentes dependendo de uma condição ser avaliada como verdadeira (diferente de `0`) ou falsa (`0`), usando `if`, `else if` e `else`.

```c
#include <stdio.h>

int main(void) {
    int nota = 15;
    
    if (nota >= 10) {
        printf("Aprovado!\n");
    } else {
        printf("Reprovado.\n");
    }
    
    return 0;
}
```

## 4. Ciclos (Loops)
Os ciclos iterativos permitem repetir instruções enquanto uma condição for verdadeira.
- `while`: A condição é testada antes de executar o bloco.
- `do ... while`: A condição é testada depois da primeira execução.
- `for`: Ideal quando o número de iterações é previamente conhecido (possui uma variável de controlo).

```c
#include <stdio.h>

int main(void) {
    printf("Ciclo for:\n");
    for (int i = 1; i <= 5; i++) {
        printf("%d ", i);
    }
    printf("\n");
    
    printf("Ciclo while:\n");
    int j = 5;
    while (j > 0) {
        printf("%d ", j);
        j--;
    }
    printf("\n");
    
    return 0;
}
```

## 5. Funções
As funções agrupam um bloco de código numa rotina reutilizável. Uma função pode receber parâmetros (argumentos) e retornar um valor. Se for chamada antes da sua definição, devemos indicar o seu `protótipo` (declaração prévia).

```c
#include <stdio.h>

// Declaracao e definicao da funcao
int somar(int a, int b) {
    return a + b;
}

int main(void) {
    int resultado = somar(10, 20);
    printf("O resultado da soma e: %d\n", resultado);
    return 0;
}
```

## 6. Arrays (Vetores)
Um array é uma coleção de elementos do **mesmo tipo**, armazenados em locais de memória contíguos. O índice do primeiro elemento é sempre `0`. Cuidado: a linguagem C não verifica automaticamente os limites do array.

```c
#include <stdio.h>

int main(void) {
    int numeros[5] = {10, 20, 30, 40, 50};
    
    for (int i = 0; i < 5; i++) {
        printf("Elemento no indice %d: %d\n", i, numeros[i]);
    }
    
    return 0;
}
```

## 7. Strings
Ao contrário de outras linguagens, o C não tem um tipo específico para strings. Elas são representadas como *arrays de caracteres* terminados pelo carácter especial nulo `\0`. A biblioteca `<string.h>` fornece funções úteis como `strlen` (tamanho) e `strcmp` (comparação).

```c
#include <stdio.h>
#include <string.h>

int main(void) {
    char saudacao[50] = "Ola, mundo!";
    
    printf("Mensagem: %s\n", saudacao);
    printf("Tamanho da string: %lu\n", strlen(saudacao));
    
    return 0;
}
```

## 8. Recursividade
A recursividade acontece quando uma função efetua chamadas a si mesma (divide and conquer). Tem sempre de ter um **caso base** que pára a recursão e o **caso recursivo** que avança para instâncias menores do problema.

```c
#include <stdio.h>

int fatorial(int n) {
    if (n == 0 || n == 1) { // Caso base
        return 1;
    }
    return n * fatorial(n - 1); // Chamada recursiva
}

int main(void) {
    int num = 5;
    printf("O fatorial de %d e: %d\n", num, fatorial(num));
    return 0;
}
```
